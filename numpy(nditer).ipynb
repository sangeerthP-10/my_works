```python
import numpy as np
```

```python
a=np.arange(12).reshape(3,4)
```

```python
print(a)
```

```python
for row in a:
  print(row)
```

```python
for row in a:
  for cell in row:
    print(cell)
```

```python
for cell in a.ravel():
  print(cell)
```

```python
for cell in a.flatten():
  print(cell)
```

```python
for cell in np.nditer(a,order='C'):
  print(cell)
```

```python
for cell in np.nditer(a,order='f'):
  print(cell)
```

```python
a
```

```python
for x in np.nditer(a,order='f', flags=['external_loop']):
  print(x)
```

```python
for x in np.nditer(a,order='f', op_flags=['readwrite']):
  x[...]=x*x
```

```python
a
```

```python
b=np.arange(3,15,4).reshape(3,1)
b
```

```python

for x,y in np.nditer([a,b]):
  print(x*y)
```

```python
sum=0
for x,y in np.nditer([a,b]):

  sum+=x*y
print(sum)
```
